The above plots states that most of the total marks of students are lying between 175-250 marks i.e, between 60-80 percentage


Assigning Grades

85-100 : Grade A
70-84 : Grade B
55-69 : Grade C
35-54 : Grade D
0-34 : Grade E

```python
def determine_grade(score):
    if score >= 85 and score <= 100:
        return ' Grade A'
    elif score >=70 and score <=84:
        return 'Grade B'
    elif score >=55 and score <=69:
        return 'Grade C'
    elif score >=35 and score <=54:
        return 'Grade D'
    elif score >=0 and score <=34:
        return 'Grade E'
```

```python
df['Grades'] = df.apply(lambda row: determine_grade(row['Percentage']), axis=1)
```
